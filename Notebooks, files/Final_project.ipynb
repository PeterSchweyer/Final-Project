```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
from selenium import webdriver
import requests
import time

url = "https://www.hochschulkompass.de/studium/studiengangsuche/erweiterte-studiengangsuche/search/1/studtyp/3.html?tx_szhrksearch_pi1%5Bresults_at_a_time%5D=100"
response = requests.get(url)
response
```

```python
index = 0
base_url = "https://www.hochschulkompass.de"
x = 2
driver = webdriver.Edge()

final_dict = {}

#try:
for i in range(220):
    driver.get(url)
    time.sleep(5)
    soup = BeautifulSoup(driver.page_source)
    response = requests.get(url)
        
    #basic
    content = soup.find_all("section", attrs = {"class":"result-box"})
    for course in content:
        Studiengang = course.find("h2").get_text()
        info = course.find_all("span", attrs = {"class":"status"})
        hochschule = info[0].get_text()
        ort = info[1].get_text()
        abschluss = info[2].get_text()
        studientyp = info[3].get_text()
        studienform = info[4].get_text()
        link = course.find("a")["href"]
        link = f"{base_url}{link}"
            
        final_dict[index] = {"Studiengang": Studiengang, 
                             "hochschule": hochschule, 
                              "ort": ort, 
                               "abschluss": abschluss, 
                               "studientyp": studientyp, 
                               "studienform": studienform, 
                               "mehr_erfahren": link}
        

        index += 1

        
    next_page = soup.find("ul", attrs = {"class":"pagination"})
    next_page = next_page.find_all("a")
    url = f"{base_url}{next_page[x]['href']}"

    if x==5:
        x = 5
    elif i<2:
        x+=1
```

```python
pd.set_option('display.max_colwidth', 200) 
df = pd.DataFrame.from_dict(final_dict, orient="index")
df
```
