```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="my_geocoder")

def get_coordinates(location):
    geocode = geolocator.geocode(location)
    if geocode:
        return geocode.latitude, geocode.longitude
    else:
        return None

locations_df['koordinaten'] = locations_df['ort'].apply(get_coordinates)
```

## Preparing for SQL/Tableu

```python
df_new = df
```

```python
df_new["abschluss"].value_counts()
```

```python
df_new["abschluss"].unique()
```

```python
patterns = {
    'Master.*': 'Master',
    'Bachelor.*': 'Bachelor',
    'Staatsprüfung.*': 'Staatsexamen',
    'Staatsexamen.*': 'Staatsexamen',
    'Diplom.*': 'Diplom',
    'Magister.*': 'Magister',
}

for pattern, replacement in patterns.items():
    df_new['abschluss'] = df_new['abschluss'].str.replace(pattern, replacement, regex=True, case=False)

df_new['abschluss'] = df_new['abschluss'].where(df_new['abschluss'].isin(['Master', 'Bachelor', 'Staatsexamen', 'Lehramt', 'Diplom', 'Magister']), 'Andere')
```

```python
df_new["abschluss"].value_counts()
```

## To SQL

```python
import pymysql
from sqlalchemy import create_engine
from getpass import getpass
password = getpass()

engine = create_engine(f'mysql+pymysql://root:{password}@127.0.0.1/studiengänge')
def save_to_mysql(df, table_name):
    df.to_sql(table_name, engine, index=False, if_exists='replace')

save_to_mysql(df_new, "df")
```
